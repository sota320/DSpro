```python
import requests

url = "https://transit.yahoo.co.jp/search/result?from=%E8%B1%8A%E6%B4%B2%E9%A7%85%E5%89%8D%2F%E9%83%BD%E5%96%B6%E3%83%90%E3%82%B9&to=%E6%AD%A6%E8%94%B5%E9%87%8E%E5%A4%A7%E5%AD%A6%E5%89%8D%2F%E9%83%BD%E5%96%B6%E3%83%90%E3%82%B9&fromgid=&togid=&flatlon=%2C%2C36054&tlatlon=%2C%2C36907&via=&viacode=&y=2024&m=12&d=23&hh=08&m1=5&m2=0&type=4&ticket=ic&expkind=1&userpass=1&ws=3&s=0&al=1&shin=1&ex=1&hb=1&lb=1&sr=1#route03"
response = requests.get(url)
```

```python
from bs4 import BeautifulSoup
import time
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
bus_stops = soup.select("ul > li > dl")
for stop in bus_stops:
    time_tag = stop.select_one("dt")
    stop_name_tag = stop.select_one("dd")
    if time_tag and stop_name_tag:
        time_info = time_tag.text.strip()
        stop_name = stop_name_tag.text.strip()
        print(f"{time_info} - {stop_name}")

time.sleep(1)
```

```python
route03_div = soup.find("div", {"id": "route03"})
if route03_div:
    bus_stops = route03_div.select("ul > li > dl")
    for stop in bus_stops:
        time_tag = stop.select_one("dt")
        stop_name_tag = stop.select_one("dd")
        if time_tag and stop_name_tag:
            time_info = time_tag.text.strip()
            stop_name = stop_name_tag.text.strip()
            print(f"{time_info} - {stop_name}")

time.sleep(1)
```
